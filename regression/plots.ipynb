# Plotting pulls and calibration

```python
import numpy as np
from plotting import make_pull_fig, make_calibration_fig
```

## In-training domain

```python
# Load datasets
data_bnn = np.load("results/bnn_results.npy", allow_pickle=True)
x_test_bnn = data_bnn.item().get("x_test")
y_test_bnn = data_bnn.item().get("y_test")
y_mean_bnn = data_bnn.item().get("mean")
y_std_bnn = data_bnn.item().get("std_total")

# GP data
data_gp = np.load("results/gp_results.npy", allow_pickle=True)
x_test_gp = data_gp.item().get("x_test")
y_test_gp = data_gp.item().get("y_test")
y_mean_gp = data_gp.item().get("mean")
y_std_gp = data_gp.item().get("std_total")

# CP data
data_cp = np.load("results/cp_results.npy", allow_pickle=True)
x_test_cp = data_cp.item().get("x_test").numpy()
y_test_cp = data_cp.item().get("y_test").numpy()
y_mean_cp = data_cp.item().get("mean").numpy()
y_std_cp = data_cp.item().get("std_total").numpy()

# RE data
data_re = np.load("results/re_results.npy", allow_pickle=True)
x_test_re = data_re.item().get("x_test")
y_test_re = data_re.item().get("y_test")
y_mean_re = data_re.item().get("mean")
y_std_re = data_re.item().get("std_total")
```

```python
make_pull_fig(
    x_values=[x_test_bnn, x_test_gp, x_test_cp, x_test_re],
    y_values=[y_test_bnn, y_test_gp, y_test_cp, y_test_re],
    y_means=[y_mean_bnn, y_mean_gp, y_mean_cp, y_mean_re],
    y_stds=[y_std_bnn, y_std_gp, y_std_cp, y_std_re],
    network_names=["BNN", "GP", "Conformal", "RE"],
    filename="results/combined_pull",
    title=r"training domain",
)
```

```python
make_calibration_fig(
    y_tests=[y_test_bnn, y_test_gp, y_test_cp, y_test_re],
    y_means=[y_mean_bnn, y_mean_gp, y_mean_cp, y_mean_re],
    y_stds=[y_std_bnn, y_std_gp, y_std_cp, y_std_re],
    network_names=["BNN", "GP", "Conformal", "RE"],
    filename="results/combined_calibration",
    title=r"training domain",
)
```
